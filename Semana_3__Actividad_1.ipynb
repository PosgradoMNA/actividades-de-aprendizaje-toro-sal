2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Estos son bases de datos gigentes que contienen informacion de diferentes  fuetes y se recopilan de acuerdo a tu nesecidad.en si permite tomar datos de muchas fuentes diferentes,

Un almacén de datos puede contener varias bases de datos. Dentro de cada base de datos, los datos se organizan en tablas y columnas. Dentro de cada columna, puede definir una descripción de los datos, como un número entero, un campo de datos o una cadena. Las tablas se pueden organizar dentro de esquemas, que puede considerar como carpetas. Cuando se ingieren datos, se almacenan en varias tablas descritas por el esquema. Las herramientas de consulta utilizan el esquema para determinar a qué tablas de datos acceder y analizar.
Este  data warehouse está especialmente diseñado para el análisis de datos, lo que implica leer grandes cantidades de datos para comprender las relaciones y las tendencias entre los datos.
